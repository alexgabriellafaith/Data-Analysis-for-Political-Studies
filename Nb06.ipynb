```python
import pandas as pd

url = "https://www.theguardian.com/news/datablog/2010/may/06/general-election-2010-turnout-since-1945"
tables = pd.read_html(url)

df = tables[0][['Year', 'UK']].copy() 

df.columns = ["Year", "Voter Turnout in UK"]
df.loc[:, "Voter Turnout in UK"] = pd.to_numeric(df["Voter Turnout in UK"], errors="coerce")

df['Year'] = df['Year'].str[:4]

df.to_csv('voter_turnout_data.csv', index=False)
```

```python
import pandas as pd
import altair as alt
from altair import datum

url = "https://www.theguardian.com/news/datablog/2010/may/06/general-election-2010-turnout-since-1945"

tables = pd.read_html(url)

df = tables[0][['Year', 'UK']].copy() 

df.columns = ["Year", "Voter Turnout in UK"]
df['Year'] = df['Year'].str[:4]

df.loc[:, "Voter Turnout in UK"] = pd.to_numeric(df["Voter Turnout in UK"], errors="coerce")

chart = alt.Chart(df).mark_line().encode(
    x='Year:O',
    y='Voter Turnout in UK:Q',
    tooltip=['Year:O', 'Voter Turnout in UK:Q']
).properties(
    title='Voter Turnout in the UK - General Elections Since 1945'
)

#chart.show('voter_turnout_chart.html')
df

```
