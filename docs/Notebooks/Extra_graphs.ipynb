# Function looking at changing economic variables overtime

This notebook does not collect data or create a new csv file it is simply an illustration of how some of the graphs in the website were created

```python
import requests
import statsmodels.api as sm
import pandas as pd
import streamlit as st
import altair as alt
from vega_datasets import data
from sqlalchemy import create_engine
from IPython.display import display
```

```python
# creating sql database engine 

engine = create_engine('sqlite:///../Data/PoliticsandDataSci.db', echo=False, isolation_level="AUTOCOMMIT")

# opening the connection to the database

def createpopulationgeo():
    
    with engine.connect() as conn:
        pass
    interactive = pd.read_sql_table("Interactive Data",con=engine)
    return interactive


database = createpopulationgeo()

```

```python
def plot_line_graph(dataframe, y_column):
    # Create Altair chart
    custom_domain = [year for year in range(1960, 2025)]
    chart = alt.Chart(dataframe).mark_line(
        color='red'
    ).encode(
        x=alt.X('Year:O', title='Year',scale=alt.Scale(domain=custom_domain)),
        y=alt.Y(f'{y_column}:Q', title=f'{y_column}'),
        tooltip=['Year:O', f'{y_column}:Q']
    ).properties(
        title=f'{y_column} Over Years'
    ).interactive()

    # Display the chart
    return display(chart)

```

```python
variables= ('Average Global Gini','UK GDP Per Capita (US $)','UK GDP Growth Rate','USA GDP Per Capita (US $)','US GDP Growth Rate')

for i in variables:
    plot_line_graph(database,i)
```
