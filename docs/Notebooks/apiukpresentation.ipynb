```python
import requests

url = "https://api.beta.ons.gov.uk/v1/datasets"

# Sending a GET request
response = requests.get(url)

# Checking the response status
if response.status_code == 200:
    # Successful request
    print("Request successful!")
    data = response.json()  # Assuming the response is in JSON format
    print(data)
else:
    # Unsuccessful request
    print(f"Error: {response.status_code}")
    print(response.text)
```

### newcastle central station 

```python


import requests

months=['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

crimerates2021newcastle=[]

for month in months:
    url = 'https://data.police.uk/api/crimes-street/all-crime?lat=54.9691552&lng=-1.6188458&date=2021-'+month+''
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json() 
        crimerates2021newcastle.append(len(data))
    else:
        print(f"Error: {response.status_code}")
        print(response.text)


print(crimerates2021newcastle)

```

```python
import requests
import pandas as pd

months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
crimerateslondon = []

for month in months:
    url = f'https://data.police.uk/api/crimes-street/all-crime?lat=51.5307106&lng=-0.1274346&date=2021-{month}'
    response = requests.get(url)
    
    if response.status_code == 200:
        # Assuming the API response is a list of dictionaries
        data_for_month = response.json()
        crimerateslondon.extend(data_for_month)

# Creating a DataFrame
crimeratemostpopular = pd.DataFrame(crimerateslondon)

# Display the DataFrame
crimeratemostpopular.to_csv('crimerates_most_popular.csv', index=False)

crimeratemostpopular
```

```python
crimecategorylondon2022 = crimeratemostpopular['category'].value_counts().reset_index()
crimecategorylondon2022.columns = ['category', 'count']


crimecategorylondon2022.to_csv('crimecategorylondon2022.csv', index=False)

crimecategorylondon2022
```

```python
import altair as alt
import pandas as pd

chart = alt.Chart(crimecategorylondon2022).mark_bar().encode(
    x='count:Q',
    y=alt.Y('category:N', sort='-x'),
    tooltip=['category:N', 'count:Q'] 
).properties(
    title='Crime by Category in London 2021'  # Add a title to the chart
)



chart
```
