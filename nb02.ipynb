```python
jsonlist = requests.get("https://api.census.gov/data/2019/pep/charagegroups?get=NAME,POP&HISP=2&for=state:*").json()
states =[]
print(jsonlist[3])
populations =[]
for list in jsonlist:
    states.append(list[0])
    populations.append(list[1])
data = {"State":states,"Population":populations
}
df=pd.DataFrame(data)
df = df.iloc[1:]
df["Population"]=pd.to_numeric(df["Population"])
df


```

```python
chart = alt.Chart(df).mark_area().encode(
    x="State",
    y="Population"
)
st.title('Population Chart')
st.altair_chart(
    chart, use_container_width=True)
!streamlit run /Users/maddonk/anaconda3/lib/python3.11/site-packages/ipykernel_launcher.py
```

```python
from vega_datasets import data

states = alt.topo_feature(data.us_10m.url, 'states')


alt.Chart(states).mark_geoshape().encode(
    color='Population:Q',
    tooltip=['State:N', 'Population:Q']
).transform_filter(
    alt.datum.State != None
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'State', ['Population'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```
